```python
#graphs
plt.plot(df['temperature'])
plt.show()
```

```python
#eda analysis and visualizaion for crop count
crop_count=df['label'].value_counts()
print(crop_count)
print(len(crop_count))
plt.figure(figsize=(12,6))
crop_count.plot(kind='bar')
plt.title('crop count analysis')
plt.xlabel('samples')
plt.ylabel('count of labels')
plt.show()
```

```python
#crop sustainability by temperature
temp_crop=df.groupby('label')['temperature'].mean()
print(temp_crop)
plt.figure(figsize=(12,6))
temp_crop.plot(kind='bar')
plt.title('samples vs temperature')

plt.xlabel('samples')
plt.ylabel('temp')
plt.show()
```

```python
#crop sustainability by temperature using scatter plot
temp_crop=df.groupby('label')['temperature'].mean()
print(temp_crop)
plt.figure(figsize=(12,6))
x=df['label']
y=df['temperature']
plt.scatter(x,y)
plt.title('samples vs temperature')

plt.xlabel('samples')
plt.ylabel('temp')
plt.show()
```

```python
#crop sustainability by humidity using scatter plot
humid_crop=df.groupby('label')['humidity'].mean()
print(humid_crop)
plt.figure(figsize=(12,6))
x=df['label']
y=df['humidity']
plt.scatter(x,y)
plt.title('label vs humidity')

plt.xlabel('label')
plt.ylabel('humidity')
plt.show()
```

```python
#crop sustainability by rainfall using scatter plot
rainfall_crop=df.groupby('label')['rainfall'].mean()
print(rainfall_crop)
plt.figure(figsize=(12,6))
x=df['label']
y=df['rainfall']
plt.scatter(x,y)
plt.title('label vs rainfall')

plt.xlabel('Label')
plt.ylabel('rainfall')
plt.show()
```
